## Precedência

```python
# exemplos de precedências dos operadores

print(2+3*2)         # primeiro multiplicação, depois soma
print((2+3)*2)       # primeiro soma, depois multiplica
print(2*5/2**2)      # primeiro exponenciação, depois multiplica, depois divide
print(2*(5/2**2))    # primeiro exponenciação, depois divisão, depois multiplica
print((2*(8/(2-1)))) # primeiro subtração, depois divisão, depois multiplica
print(8/2*2)         # primeiro divisão, depois multiplica
print(2**2**3)       # primeiro a exponenciação mais a direita, depois a esquerda
```

Exercícios:
1) Conversor de Temperatura: temperatura em Celsius. Ex: 25ºc Converta para Fahrenheit usando a fórmula: (0 °C * 9/5) + 32 = 32 °F Imprima a temperatura em Fahrenheit.

2) Desconto: Informe o valor de um produto e o percentual de desconto. Calcule o valor do desconto e o preço final. Exiba os resultados.

3) Velocidade Média: Informe a distância percorrida e o tempo gasto. Calcule a velocidade média usando a fórmula: VM = Distância / Tempo

```python
# Exercicio 1: conversor de temperatura de Celsius para Fahrenheit:

temp_celsius = float(input('Digite a temperatura em Celsius: '))

temp_fahrenheit = (temp_celsius * 9/5) + 32

print(f'A temperatura em Fahrenheit é: {temp_fahrenheit:.2f}')
```

```python
# Exercicio 2: calculadora de valor com desconto:

valor_original = float(input('Digite o valor do produto: '))
desconto = float(input('Digite o percentual de desconto: '))/100

print(f'O valor do desconto é: {valor_original * (desconto):.2f}%')
print(f'o valor final é: {valor_original - (valor_original * (desconto)):.2f}')
```

```python
# Exercicio 3: calculo de velocidade média

distancia = float(input('Digite a distância percorrida: '))
tempo = float(input('Digite o tempo gasto: '))

velocidade_media = distancia / tempo

print(f'Sua velocidade média foi: {velocidade_media}')
```

```python
# Exercicio 4:
# Volume do Cubo.
# Solicite a aresta do cubo.
# calcule o volume usando a dormula: volume = aresta³.

aresta_cubo = float(input('Digite a aresta do cubo: '))
volume_cubo = aresta_cubo**3

print(f'O volume deste cubo é: {volume_cubo:.2f}cm³')
```

```python
# Exercicio 5:
# Solicite o ano de nascimento.
# Calcule a idade considerando o ano atual
# Exiba a idade

ano_atual = 2026
ano_nascimento = int(input('Digite o seu ano de nascimento: '))
idade = ano_atual - ano_nascimento

print(f'Você tem {idade} anos')
```

```python
# Exercicio 6:
# Tabuada: Solicite um número.
# Imprima a tabuada desse número até 10.

numero = int(input('Digite um número: '))

print('')

print(f'A tabuada do numero {numero} é: ')
for i in range(1,11):
    print(f'{numero} x {i} = {numero * i}')
```
