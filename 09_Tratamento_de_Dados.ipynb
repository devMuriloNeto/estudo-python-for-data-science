```python
# alterando o tipo de dado de string (object) para datetime (ao trazer o arquivo)

df_vendas = pd.read_csv('/content/drive/MyDrive/!Data Python 2604/base_vendas.csv', parse_dates=['dt_venda'], dayfirst=True)
df_vendas.head(10)
```

```python
# alterando o tipo de dado de string (object) para datetime (maior controle)

df_vendas['dt_entrega'] = pd.to_datetime(df_vendas['dt_entrega'], format='%d/%m/%Y %H:%M')
```

```python
# trocar a vírgula por ponto no campo 'object' (campo = preco_custo)

df_vendas['preco_custo'] = df_vendas['preco_custo'].str.replace(',','.')
df_vendas.head()
```

```python
# alterar do tipo string para float

df_vendas['preco_custo'] = df_vendas['preco_custo'].astype(float)
df_vendas.info()
```

```python
# trocar a vírgula por ponto no campo 'object' e alterar para float (campo = valor_unitario)

df_vendas['valor_unitario'] = df_vendas['valor_unitario'].str.replace(',','.').astype(float)
df_vendas.info()
```

```python
# remover colunas indesejadas na análise

df_vendas.drop(columns=['matricula_funcionario','codigo_produto'], inplace=True)
df_vendas
```

```python
# ENGENHARIA DE DADOS / DATA ENGINEER
# Criando novas colunas
# 1) Valor Final

df_vendas['valor_final'] = (df_vendas['quantidade'] * df_vendas['valor_unitario']).round(2)

# df_vendas.insert(9, 'valor_final', df_vendas['quantidade'] * df_vendas['valor_unitario'])
df_vendas
```

```python
# 2) Valor Comissao

df_vendas['valor_comissao'] = (df_vendas['valor_final'] * df_vendas['comissao']).round(2)
df_vendas
```

```python
# 3) Lucro Bruto

df_vendas['lucro_bruto'] = (df_vendas['valor_unitario'] - df_vendas['preco_custo']) * df_vendas['quantidade']
df_vendas
```

```python
# 4) Lucro Real

df_vendas['lucro_real'] = df_vendas['lucro_bruto'] - df_vendas['valor_comissao']
df_vendas
```

```python
# 5) Margem de Lucro

df_vendas['margem_lucro'] = (((df_vendas['valor_unitario'] - df_vendas['preco_custo']) / df_vendas['valor_unitario'])).round(2) * 100
df_vendas
```
