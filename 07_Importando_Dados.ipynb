```python
## Instalar o conector MySQL e Pandas
!pip install mysql-connector-python pandas


## Importar bibliotecas
import mysql.connector
import pandas as pd


# conexão com o banco de dados MySQL
conexao = mysql.connector.connect(
    host="banco2604a.mysql.dbaas.com.br",
    user="banco2604a",
    password="Senai@2604",
    database="banco2604a"
)
print('👍Conexão estabelecida com sucesso\n')


## Criar cursor com dicionário
cursor = conexao.cursor(dictionary=True)


## 1️⃣ Pedidos de cada cliente
query = """
SELECT
    c.id_cliente,
    c.nome,
    c.email,
    p.id_pedido,
    pr.nome AS produto,
    p.quantidade,
    p.status,
    ce.cidade,
    ce.estado
FROM pedidos AS p
INNER JOIN clientes AS c ON p.id_cliente = c.id_cliente
INNER JOIN produtos AS pr ON p.id_produto = pr.id_produto
INNER JOIN ceps AS ce ON c.cep = ce.cep
ORDER BY c.nome;
"""
cursor.execute(query)
pedidos_clientes = pd.DataFrame(cursor.fetchall())
print("\n\n📌 Pedidos de cada cliente:")
display(pedidos_clientes)
```
