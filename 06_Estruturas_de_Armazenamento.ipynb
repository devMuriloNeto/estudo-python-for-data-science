```python
# Crie um programa que vai ler vários números e colocar em uma lista.
# Depois disso, mostre:
# A uma lista só com pares
# B uma lista só com ímpares
# C uma lista com todos os números

pares = []
impares = []
lista = []

while True:
    num = input('Digite um número: ("S" para Sair): ').lower()
    if num == 's':
        break
    else:
        num = int(num)
        lista.append(num)
        if num % 2 == 0:
            pares.append(num)
        else:
            impares.append(num)

lista.sort()
pares.sort()
impares.sort()

print('Números digitados:', lista)
print('Números pares:', pares)
print('Números impares:', impares)
```

```python
# recriar o exercício Maior Número / Menor Número usando listas

numeros = []

while True:
    num = input('Digite um número ("S" para Sair): ').lower()

    if num == 's':
        break
    else:
        num = float(num)
        numeros.append(num)
media = sum(numeros)/len(numeros)
print(f'Maior: {max(numeros)}')
print(f'Menor: {min(numeros)}')
print(f'Soma:  {sum(numeros)}')
print(f'Quant: {len(numeros)}')
print(f'Média: {media}')
```

```python
# CRIAR UM CAMPEONATO;

# CRIAR AUTOMATICAMENTE OS RESULTADOS DOS JOGOS!
# CADA TIME ENFRENTA UM TIME DO OUTRO GRUPO EM JOGO ÚNICO

# JOGO 16
# JOGO 8
# JOGO 4
# SEMI
# FINAL

"""
Dada a lista com 32 Times, faça o sorteio para outras duas listas (A e B).
Ao final, cada nova lista deve conter 16 times.
random.choice(times)
"""
from random import choice
from random import randint

times=["Flamengo", "Palmeiras", "São Paulo", "Athletico-PR", "Atlético-MG" , "Corinthians", "Fluminense", "Grêmio", "Fortaleza", "Internacional", "Bahia", "Botafogo", "Red Bull Bragantino", "Atlético-GO", "Ceará", "Cuiabá", "Goiás", "Vasco", "Juventude", "Sport", "CRB", "Vitória", "Criciúma", "Sampaio Corrêa", "Operário-PR" , "Botafogo-SP", "Brusque", "Ypiranga-RS", "América-RN", "Amazonas", "Águia de Marabá", "Sousa-PB"]

times_a = []
times_b = []

for i in range(1,17):
    sorteado = choice(times)
    times_a.append(sorteado)
    times.remove(sorteado)

    sorteado = choice(times)
    times_b.append(sorteado)
    times.remove(sorteado)

print(f'Times Grupo A: \n {times_a}')
print(f'Times Grupo B: \n {times_b}')

############################################

print('\n\nJOGOS 16 AVOS')

print('\nCHAVE A \n')

x = 0
while x < 8:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)

        time1 = choice(times_a)
        times_a.remove(time1)
        time2 = choice(times_a)

        if (golsA != golsB):
            break

    print(f'{time1} {golsA} x {golsB} {time2}')
    if golsA < golsB:
        times_a.remove(time1)
    else:
        times_a.remove(time2)
    x += 1

print('\nCHAVE B \n')

x = 0
while x < 8:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_b[x] } {golsA} x {golsB} {times_b[x+1]}')
    if golsA < golsB:
        times_b.remove(times_b[x])
    else:
        times_b.remove(times_b[x+1])
    x += 1

print(f'\nTimes Grupo A: \n {times_a}')
print(f'Times Grupo B: \n {times_b}')

############################################

print('\n\nJOGOS OITAVAS')

print('\nCHAVE A \n')

x = 0
while x < 4:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_a[x] } {golsA} x {golsB} {times_a[x+1]}')
    if golsA < golsB:
        times_a.remove(times_a[x])
    else:
        times_a.remove(times_a[x+1])
    x += 1

print('\nCHAVE B \n')

x = 0
while x < 4:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_b[x] } {golsA} x {golsB} {times_b[x+1]}')
    if golsA < golsB:
        times_b.remove(times_b[x])
    else:
        times_b.remove(times_b[x+1])
    x += 1

print(f'\nTimes Grupo A: \n {times_a}')
print(f'Times Grupo B: \n {times_b}')

############################################

print('\n\nJOGOS QUARTAS')

print('\nCHAVE A \n')

x = 0
while x < 2:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_a[x] } {golsA} x {golsB} {times_a[x+1]}')
    if golsA < golsB:
        times_a.remove(times_a[x])
    else:
        times_a.remove(times_a[x+1])
    x += 1

print('\nCHAVE B \n')

x = 0
while x < 2:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_b[x] } {golsA} x {golsB} {times_b[x+1]}')
    if golsA < golsB:
        times_b.remove(times_b[x])
    else:
        times_b.remove(times_b[x+1])
    x += 1

print(f'\nTimes Grupo A: \n {times_a}')
print(f'Times Grupo B: \n {times_b}')

############################################

print('\n\nJOGOS SEMI-FINAL')

print('\nCHAVE A \n')

x = 0
while x < 1:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_a[x] } {golsA} x {golsB} {times_a[x+1]}')
    if golsA < golsB:
        times_a.remove(times_a[x])
    else:
        times_a.remove(times_a[x+1])
    x += 1

print('\nCHAVE B \n')

x = 0
while x < 1:
    while True:
        golsA = randint(0, 5)
        golsB = randint(0, 5)
        if golsA != golsB:
            break
    print(f'{times_b[x] } {golsA} x {golsB} {times_b[x+1]}')
    if golsA < golsB:
        times_b.remove(times_b[x])
    else:
        times_b.remove(times_b[x+1])
    x += 1

print(f'\nTimes Grupo A: \n {times_a}')
print(f'Times Grupo B: \n {times_b}')

############################################

print('\n\nJOGO FINAL')

while True:
    golsA = randint(0, 5)
    golsB = randint(0, 5)
    if golsA != golsB:
        break

print(f'{times_a[0] } {golsA} x {golsB} {times_b[0]}')

if golsA < golsB:
    print(f'\nCampeão: {times_b[0]} 🏆')
else:
    print(f'\nCampeão: {times_a[0]} 🏆')

###########################################

print('\n\nJOGO FINAL')

while True:
    golsA = randint(0, 5)
    golsB = randint(0, 5)
    if golsA != golsB:
        break

print(f'{times_a[0] } {golsA} x {golsB} {times_b[0]}')

if golsA < golsB:
    print(f'\nCampeão: {times_b[0]} 🏆')
else:
    print(f'\nCampeão: {times_a[0]} 🏆')
```

```python
# Estruturas de Armazenamento --> TUPLAS

# VARIÁVEL = armazena um valor único e um único tipo de dado;
# LISTAS   = armazena N valores e N tipos de dados; # Mutável
# TUPLAS   = armazena N valores e N tipos de dados; # Imutável

```

```python
tupla = (12, True, 'Senai', 3.14)
print(tupla)
print(type(tupla))
```

```python
tupla2 = tuple([15,16,'senai',15.89])
print(tupla2)
print(type(tupla2))
```

```python
tupla_bancos = 'Itaú', 'Santander', 'Bradesco', 'Caixa', 'Inter', 'Safra', 'NuBank', 'C6 Bank', False, 1, ['a','b','c']
print(tupla_bancos)
print(type(tupla_bancos))
print(len(tupla_bancos))
```

```python
# exibir itens da tupla

print(tupla_bancos[0][0])
print(tupla_bancos[0][0:2])
print(tupla_bancos[-1])
print(tupla_bancos[1:3])
print(tupla_bancos[3:7])
print(tupla_bancos[7:])
print(tupla_bancos[10][2])
```

```python
# exibir a posição de um elemento na tupla

print(tupla_bancos.index('Safra'))
```

```python
# não é possível alterar elementos de uma tupla (exceto um item de uma lista dentro da tupla)

tupla_bancos[2] = 'Crefisa'
# del tupla_bancos[1:5]
```

```python
# é possível alterar um item de uma lista dentro da tupla

tupla_bancos[10][2] = 'd'

print(tupla_bancos)
```

```python
tupla_bancos.append('Will Bank')
```

```python
# pesquisar item na tupla

print('Itaú' in tupla_bancos)
print('Itaú' not in tupla_bancos)
```

```python
# listar elementos da tupla

print(tupla_bancos)

for i in tupla_bancos:
    print(i)

for i in tupla_bancos:
    print(i, end=', ')
print('\b\b.')
```

```python
tupla_imutavel = (1, 2, 3)
lista_mutavel = list(tupla_imutavel)

print(type(tupla_imutavel))
print(type(lista_mutavel))

lista_mutavel.append(4)
lista_mutavel.append(5)
lista_mutavel.extend([6,7,8,9,10])

print(lista_mutavel)

tupla_imutavel = tuple(lista_mutavel)

print(type(tupla_imutavel))
print(type(lista_mutavel))

print(tupla_imutavel)
print(lista_mutavel)
```
