```python
# #Desenvolva um programa que leia dez números inteiros e mostre a soma apenas
# daqueles que forem pares. Se o valor digitado for ímpar desconsidere-o.
# contar os números pares digitados.

soma_pares = 0
qtd_pares = 0

for i in range(1,11):
    numero = int(input(f'Digite o {i}º numero: '))

    if numero % 2 == 0:
        soma_pares += numero
        qtd_pares += 1


print(f'\na soma dos numeros pares é: {soma_pares}\n')
print(f'a quantidade dos numeros pares é: {qtd_pares}')
```

```python
# Mostre a tabuada de um número que o usuário escolher.

numero = int(input('Digite um numero para ver a tabuada: '))

print(f'\nA tabuada do numero {numero} é: ')
for i in range(1,11):
    print(f'{numero} X {i} = {numero*i}')
```

```python
# Crie um programa que mostre na tela todos os números pares que estão
# no intervalo entre 1 e 50 na horizontal

for i in range(1,51):
    if i % 2 == 0:
        print(i, end=', ')
print('\b\b.')
```

```python
# Faça um programa que leia o peso de 12 pessoas.
# No final, mostre qual foi o maior e o menor peso lidos.

peso_maior = 0
peso_menor = 0

for i in range(1,13):
    peso = float(input(f'Digite o peso da {i}ª pessoa: '))

    if peso_maior == 0 or peso > peso_maior:
            peso_maior = peso
    if peso_menor == 0 or peso < peso_menor:
            peso_menor = peso

print(f'\nO maior peso foi: {peso_maior}kg\n')
print(f'o menor peso foi: {peso_menor}kg\n')
```

```python
# Faça um programa que peça a idade de 15 pessoas
# mostre quantas são maiores de idade e quantas são menores de idade

qtd_maiores = 0
qtd_menores = 0

for i in range(1,16):
    idade = float(input(f'Digite a idade da {i}ª pessoa: '))

    if idade < 18:
        qtd_menores += 1
    else:
        qtd_maiores += 1

print(f'A quantidade de pessoas menores de idade é: {qtd_maiores}')
print(f'A quantidade de pessoas maiores de idade é: {qtd_menores}')
```
