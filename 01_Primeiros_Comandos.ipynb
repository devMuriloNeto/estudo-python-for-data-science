```python
# monte um script que receba o peso e a altura de uma pessoa e mostre o IMC

# IMC = peso dividido pela altura ao quadrado
```

```python
# Solicita o peso ao usuário
peso = float(input('Digite o seu peso em kg (ex: 70.5): '))

# Solicita a altura ao usuário
altura = float(input('Digite a sua altura em metros (ex: 1.75): '))

# Calcula o IMC
# IMC = peso / (altura * altura)
imc = peso / (altura ** 2)

# Exibe o resultado do IMC
print(f'Seu IMC é: {imc}')
```
